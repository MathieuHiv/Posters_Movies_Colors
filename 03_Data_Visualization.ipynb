```python
df_plot = table_color[table_color["Main Color"] == "Main Color"].groupby(["Année", "Base_Color"] )["Durée"].count().reset_index(name='Count').sort_values(['Count'], ascending=False)
dfyear = df_plot["Année"].values.tolist()
dfcolor = df_plot["Base_Color"].values.tolist()
year = []
color = []
for i in range(len(dfyear)):
    if dfyear[i] not in year:
        year.append(dfyear[i])
        color.append(dfcolor[i])
color_year = pd.DataFrame(columns = ["Year","Col"])
color_year["Year"] = year
color_year["Col"] = color

year = table_color.groupby("Année")["Durée"].count().reset_index(name='Count').sort_values(['Année'], ascending=False)
color_year.sort_values(['Year'], ascending=False, inplace = True)
colors = color_year["Col"].values.tolist()

fig, ax= plt.subplots(figsize = (15, 8))
ax.scatter(year["Année"], year["Count"], c = colors, s = 50 )
plt.title('Most Main Color used for a movie poster per year', size=16)

plt.savefig('Most_Main_Color.png', bbox_inches ="tight")

plt.show()
```

```python
for df, color, count in zip(
    [df_black,df_cyan,df_gray,df_purple, df_brown, df_blue, df_pink, df_orange, df_red, df_yellow, df_green, df_white],
    ['black', 'cyan', 'gray', 'purple', 'brown', 'blue', 'pink', 'orange','red', 'yellow', 'green', 'white'],
    ['count_black', 'count_cyan', 'count_gray', 'count_purple', 'count_brown', 'count_blue', 'count_pink', 'count_orange','count_red', 'count_yellow', 'count_green', 'count_white']):
    df['year'] = df.iloc[:,0]
    df['color'] = [color] * len(df)
    df['count'] = df[count]
    df = df[['year', 'color', 'count']]

bar_chart_df = pd.concat([df_black,df_cyan,df_gray,df_purple, df_brown,df_blue, df_pink, df_orange, df_red, df_yellow, df_green, df_white], ignore_index=True, axis=0, sort=False)[['year', 'color', 'count']]

df_bc = pd.DataFrame(index=['black', 'cyan', 'gray', 'purple', 'brown', 'blue', 'pink', 'orange','red', 'yellow', 'green', 'white'])
for year,color,count in zip(bar_chart_df['year'],bar_chart_df['color'],bar_chart_df['count']):
    df_bc.loc[color, year] = count
df_bc = df_bc.fillna(0)


df_bc
```
